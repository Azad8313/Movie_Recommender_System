```python
import ast

def convert(obj):
    L=[]
    for i in  ast.literal_eval(obj):    #Convert string to list
        L.append(i['name'])
    return L    
```

```python
df3['genres']=df3['genres'].apply(convert)
```

```python
df3.head()
```

```python
# convert 'keywords' column to right format
df3.iloc[0].keywords
```

```python
df3['keywords']=df3['keywords'].apply(convert)
```

```python
df3.head()
```

```python
#Convert 'cast' column in right format upto 3.
def convert3(obj):
    l=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            l.append(i['name'])
            count+=1
        else:
            break
    return l       
```

```python
df3['cast']=df3['cast'].apply(convert3)
```

```python
df3.head()
```

```python
df3['crew'][0]
```

```python
# Fetch Directer name from 'crew' column. it is usefull for recomendation system
import ast

def fetch_director(obj):
    L=[]
    for i in  ast.literal_eval(obj):    #Convert string to list
        if i['job']== 'Director':
            L.append(i['name'])
            break
    return L  
```

```python
df3['crew']=df3['crew'].apply(fetch_director)
```

```python
df3.head()
```

```python
df3['overview'][0]
```

```python
# Convert 'overview' column string to list.
df3['overview']=df3['overview'].apply(lambda x: x.split())
```

```python
df3.head()
```

**Removing space between name or keyword 'like Abdul' Azad will be 'AbdulAzad'**

```python
df3['genres']=df3['genres'].apply(lambda x:[i.replace(" ","")for i in x])
df3['keywords']=df3['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
df3['cast']=df3['cast'].apply(lambda x:[i.replace(" ","")for i in x])
df3['crew']=df3['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
df3.head()
```

```python
# Concatenate the column.
df3['tags']=df3['overview']+ df3['genres']+ df3['keywords']+ df3['cast']+ df3['crew']
```

```python
df3.head()
```

```python
df4=df3[['movie_id','title','tags']]
df4.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

# s='loved loving love'
# def c(s):
#     a=[]
#     for i in s.split():
#         a.append(ps.stem(i))
#         print(a)
#     return " ".join(a)

# b=c(s)
# print(b)
```

```python
# Streming the words. Streming is like ['loved','loving','loved'] = ['love','love','love'].
def strem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
df4['tags']=df4['tags'].apply(strem)
```

```python
# Converting tags column list to string
df4['tags']=df4['tags'].apply(lambda x:" ".join(x))
```

```python
type(df4['tags'][0])
```

```python
df4['tags'][0]
```

```python
# Converting all tags to lowercase.
df4['tags']=df4['tags'].apply(lambda x: x.lower())
```

```python
df4.tags[0]
```

```python
# Vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(df4['tags']).toarray()
vectors
```

```python
vectors.shape
```

```python
vectors[0]    #Avtar
```

```python
cv.get_feature_names()
```

```python
# Calculate cosine_similarity between every vector.
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors).shape
```

```python
similar=cosine_similarity(vectors)
print(similar.shape)
similar                  # 1 means similar and 0 means opposite.
```

```python
similar[0]
```

```python
def recomender(movie):
    movie_index=df4[df4['title']==movie].index[0]
    distances=similar[movie_index]
    movie_list=sorted(enumerate(distances),reverse=True,key=lambda x: x[1])[1:8]
    
    for i in movie_list:
        print(df4.iloc[i[0]]['title'])     #title or 1 is same for fetching movie name
```

```python
recomender('Avatar')
```
